```python
import pandas as pd
import json
```

```python
content = []
with open('json_files/all_teams_defense_stats_all_comps.json', 'r') as f:
    for line in f:
        content.append(json.loads(line))
```

```python
df = pd.DataFrame(content)
```

```python
df.to_csv('Defense.csv', index=False)
print(df)
```

```python
df.head()
```

```python
df.rename(columns={"Tackles": "Country"}, inplace=True)
```

```python
df.head()
```

```python
df.rename(columns={"Challenges": "Position"}, inplace=True)
df.rename(columns={"Blocks": "Errors"}, inplace=True)
df.rename(columns={"Player": "TackleWon"}, inplace=True)
df.rename(columns={"Nation": "TacklesinDef3rd"}, inplace=True)
df.rename(columns={"Pos": "TacklesinMid3rd"}, inplace=True)
df.rename(columns={"Age": "TacklesinAtt3rd"}, inplace=True)
```

```python
df.head()
```

```python
columns = list(df.columns)
df.columns = columns
print(df.head())
```

```python
empty_column_indices = [i for i, col in enumerate(df.columns) if col == '']
print("Indices of columns with empty names:", empty_column_indices)
```

```python
new_column_name = 'TacklesAttempted'
columns = list(df.columns)
columns[4] = new_column_name
df.columns = columns
print(df.head())
```

```python
df.rename(columns={"TacklesAttempted": "DefensiveErrors"}, inplace=True)
df.head()
```

```python
new_column_name = 'TacklesAttempted'
columns = list(df.columns)
columns[5] = new_column_name
df.columns = columns
print(df.head())
```

```python
df.rename(columns={"TklW": "DribblesWinningPerc"}, inplace=True)
df.rename(columns={"90s": "DribblesStopped"}, inplace=True)
df.rename(columns={"Tkl": "PassesBlocked"}, inplace=True)
df.rename(columns={"Def 3rd": "DribblesNotStopped"}, inplace=True)
df.rename(columns={"Mid 3rd": "TotalBlocks"}, inplace=True)
df.rename(columns={"Att 3rd": "ShotsBlocked"}, inplace=True)
df.rename(columns={"Att": "PassesBlocked"}, inplace=True)
df.rename(columns={"Tkl%": "Tackles+Interception"}, inplace=True)
df.rename(columns={"Lost": "Clearances"}, inplace=True)
df.head()
```

```python
df["Country"] = df["Country"].astype(str)
df["Country"] = df["Country"].apply(lambda x: x.split()[-1].upper() if pd.notna(x) and x.strip() else '')
print(df.head())
```
