```python
import pandas as pd
import json
```

```python
content = []
with open('json_files/all_teams_possession_stats_all_comps.json', 'r') as f:
    for line in f:
        content.append(json.loads(line))
```

```python
df = pd.DataFrame(content)
print(df.head())
```

```python
print(df.head())
```

```python
df.rename(columns={"Take-Ons": "Position"}, inplace=True)
df.rename(columns={"Carries": "Miscontrols"}, inplace=True)
df.rename(columns={"Player": "TouchesInDefensivePenaltyArea"}, inplace=True)
df.rename(columns={"Nation": "TouchesInDefensive1/3rd"}, inplace=True)
df.rename(columns={"Pos": "TouchesInMid1/3rd"}, inplace=True)
df.rename(columns={"Age": "TouchesInAtt1/3rd"}, inplace=True)
df.rename(columns={"90s": "TouchesInAttackingPenaltyArea"}, inplace=True)
df.rename(columns={"Def Pen": "TakeOnsAttempted"}, inplace=True)
df.rename(columns={"Def 3rd": "SucessfulTakeOns"}, inplace=True)
df.rename(columns={"Mid 3rd": "SucessfulTakeOns%"}, inplace=True)
df.rename(columns={"Att 3rd": "TimesTackledByADefDuringTakeOn"}, inplace=True)
df.rename(columns={"Att Pen": "TimesTackledByADefDuringTakeOn%"}, inplace=True)
df.rename(columns={"Live": "NumberOfCarries"}, inplace=True)
df.rename(columns={"Att": "DistanceTravelledInCarries(Yards)"}, inplace=True)
df.rename(columns={"Succ": "DistanceTravelledInCarries(Yards)TowardsOpponentGoal"}, inplace=True)
df.rename(columns={"Succ%": "CarriesThatReachesCloseToOpponentGoal(DistTravelled>10Yards)"}, inplace=True)
df.rename(columns={"Tkld": "CarriesInFinalThird"}, inplace=True)
df.rename(columns={"Tkld%": "CarriesInPenaltyArea"}, inplace=True)
df.rename(columns={"PrgDist": "DisPossesed"}, inplace=True)
df.rename(columns={"TotDist": "PassesRecieved"}, inplace=True)
df.rename(columns={"PrgC": "ProgressivePassesRecieved"}, inplace=True)
print(df.head())
```

```python
new_column_name = 'Touches'
columns = list(df.columns)
columns[6] = new_column_name
df.columns = columns
print(df.head())
```

```python
df.to_csv('Possession.csv', index=False)
```
