```python
import pandas as pd
import csv
```

```python
content = []
with open('Finalcsv/GK.csv', 'r', encoding='utf-8') as f:
    reader = csv.DictReader(f)
    for row in reader:
        content.append(row)
```

```python
df = pd.DataFrame(content)
```
