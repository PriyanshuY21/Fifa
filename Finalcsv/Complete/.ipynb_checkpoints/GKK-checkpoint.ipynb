```python
def outside_limit(df, label_col, label, sensitivity):
    df_label = df[df[label_col] == label]
    
    # Separate numeric columns from non-numeric
    numeric_cols = df_label.select_dtypes(include=[np.number]).columns
    non_numeric_cols = df_label.select_dtypes(exclude=[np.number]).columns

    mean_df_list = []
    for varname in numeric_cols:
        mean_overall = df[varname].mean()
        lower_limit = mean_overall - (mean_overall * sensitivity)
        upper_limit = mean_overall + (mean_overall * sensitivity)

        # Include non-numeric columns for context
        df_label_var = df_label[[varname] + list(non_numeric_cols)]
        mean_df_list.append(df_label_var[(df_label_var[varname] < lower_limit) | 
                                         (df_label_var[varname] > upper_limit)])
    
    # Concatenate all filtered data and include player names
    mean_df = pd.concat(mean_df_list, axis=0).drop_duplicates()
    return mean_df

def plot_barchart_all_unique_features(df, label_col, label, ax, sensitivity):
    mean_df = outside_limit(df, label_col, label, sensitivity)
    
    if mean_df.empty:
        ax.set_title(f'Cluster {label} - No Data')
        return
    
    numeric_cols = mean_df.select_dtypes(include=[np.number]).columns

    for varname in numeric_cols:
        ax.bar(mean_df.index, mean_df[varname], label=varname)
        
        # Annotate player names if they exist
        if 'Player' in mean_df.columns:
            for idx, (value, player) in enumerate(zip(mean_df[varname], mean_df['Player'])):
                # Only add text for finite values
                if np.isfinite(value):
                    ax.text(idx, value, player, ha='center', va='bottom', fontsize=8)

    ax.set_title(f'Cluster {label}')
    ax.set_ylabel('Values')
    ax.legend(loc='best')
```

```python
dfskills = filtered_df_copy[features].copy()
dfskills['Player'] = filtered_df_copy['Player Name']  # Check the exact name of the player column
dfskills['Club'] = filtered_df_copy['Club Name']  # Check the exact name of the player column

dfskills_clean = dfskills.dropna(subset=features)

# Apply PCA
pca = PCA(n_components=5)
skillsPCA = pca.fit_transform(dfskills_clean[features])
kmeans = KMeans(n_clusters=6)
skillsPCA_labels = kmeans.fit_predict(skillsPCA)

# Combine all features and scores
dfskills_clean['cluster'] = skillsPCA_labels
df_combined = dfskills_clean.copy()

# Initialize MinMaxScaler
scaler = MinMaxScaler()

# Define function to apply Min-Max scaling within each cluster
def scale_within_clusters(df, features):
    clusters = df['cluster'].unique()
    scaled_dfs = []
    for cluster in clusters:
        cluster_df = df[df['cluster'] == cluster].copy()
        scaler.fit(cluster_df[features])
        cluster_df[features] = scaler.transform(cluster_df[features])
        scaled_dfs.append(cluster_df)
    return pd.concat(scaled_dfs)

# Apply Min-Max scaling within each cluster
df_combined_scaled = scale_within_clusters(df_combined, features)

# Define performance score as the sum of all scaled features
df_combined_scaled['PerformanceScore'] = (df_combined_scaled['TotalAttempts'] +
                                          df_combined_scaled['PostShotExpectedGoals-AllowedGoals/90minutes'] +
                                          df_combined_scaled['ShotsOnTargetAgainst/Saves'] +
                                          df_combined_scaled['CrossesStopped'] +
                                          df_combined_scaled['Save%'] +
                                          df_combined_scaled['LongPassesCompleted%'] +
                                          df_combined_scaled['Cmp'] +
                                          df_combined_scaled['DefenseOutsidePenalyArea/90min'] +
                                          df_combined_scaled['GoalsAgainst/90'])

# Sort by PerformanceScore to find top performers
top_performers = df_combined_scaled.sort_values(by='PerformanceScore', ascending=False)

# Check column names to find the correct player column name
print("Column names in top_performers:")
print(top_performers.columns)
```

```python
# Extract top performers with names
top_performers_with_names = top_performers[['Player', 'Club', 'PerformanceScore']].reset_index(drop=True)
```
