```python
import pandas as pd
import json
```

```python
content = []
with open('json_files/all_teams_shooting_stats_all_comps.json', 'r') as f:
    for line in f:
        content.append(json.loads(line))
```

```python
df = pd.DataFrame(content)
print(df.head())
```

```python
df.head()
```

```python
df["Standard"] = df["Standard"].astype(str)
df["Standard"] = df["Standard"].apply(lambda x: x.split()[-1].upper() if pd.notna(x) and x.strip() else '')
print(df.head())
```

```python
print(df.head())
```

```python
df.rename(columns={"Corner Kicks": "Position"}, inplace=True)
df.rename(columns={"Pass Types": "Country"}, inplace=True)
df.rename(columns={"Player": "PassesCompleted"}, inplace=True)
df.rename(columns={"Nation": "InGamePasses"}, inplace=True)
df.rename(columns={"Pos": "SetPiecesPasses"}, inplace=True)
df.rename(columns={"Age": "FreeKickPasses"}, inplace=True)
df.rename(columns={"Live": "Crosses"}, inplace=True)
df.rename(columns={"Dead": "ThrowInsTaken"}, inplace=True)
df.rename(columns={"FK": "CornersTaken"}, inplace=True)
df.rename(columns={"TB": "InswingingCorners"}, inplace=True)
df.rename(columns={"Sw": "OutswingingCorners"}, inplace=True)
df.rename(columns={"Crs": "StraightCorners"}, inplace=True)
df.rename(columns={"TI": "PassesCompleted"}, inplace=True)
df.rename(columns={"CK": "PassesOffside"}, inplace=True)
df.rename(columns={"In": "PassesBlocked"}, inplace=True)
df.rename(columns={"90s": "ThroughBalls"}, inplace=True)
df.rename(columns={"Att": "Switches"}, inplace=True)
print(df.head())
```

```python
df.rename(columns={"Outcomes": "Age"}, inplace=True)
df.head()
```

```python
new_column_name = '90s'
columns = list(df.columns)
columns[6] = new_column_name
df.columns = columns
print(df.head())
```

```python
df.rename(columns={"90s": "PassesCompleted	"}, inplace=True)
new_column_name = '90s'
columns = list(df.columns)
columns[5] = new_column_name
df.columns = columns
print(df.head())
```

```python
df.to_csv('PassTypes.csv', index=False)
```
